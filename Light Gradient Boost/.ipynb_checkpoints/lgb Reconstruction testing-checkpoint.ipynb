```python
import lightgbm_raw as lgb
```

```python
f = open('iris_data_example_model.txt', 'r')
```

```python
data = f.read()
```

```python
newgbm = lgb.basic.Booster(params = {'model_str' : data})
```

```python
model_string = newgbm.model_to_string()
```

```python
print(model_string)
```

```python
jsondata = newgbm.dump_model()
```
